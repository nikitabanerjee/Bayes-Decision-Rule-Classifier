```python
centroid_X1=np.mean(X1, axis=1)
print(X1)
centroid_X2=np.mean(X2, axis=1)
print(X2)
```

```python
def euclidean_distance(p,q):
    distance=0
    p=np.asarray(p)
    q=np.asarray(q)
    
    for i in range(2):
        distance = distance + (np.square(p[i] - q[i]))
                                   
    return np.sqrt(distance)
```

```python
classification_error=0
for x in Y1:
    distance1=euclidean_distance(x, centroid_X1)
    distance2=euclidean_distance(x,centroid_X2)
    if(distance1>distance2):
        classification_error+=1
```

```python
classification_error2=0
for x in Y2:
    distance1=euclidean_distance(x, centroid_X1)
    distance2=euclidean_distance(x,centroid_X2)
    if(distance1<distance2):
        classification_error2+=1
```

```python
z=(classification_error)/2500
print(z)
```

```python
m1 = [0,0,0,0,0]
m2 = [1,1,1,1,1]
S1 = [[0.8, 0.2, 0.1, 0.05, 0.01],[0.2, 0.7, 0.1, 0.03, 0.02],[0.1, 0.1, 0.8, 0.02, 0.01], [0.05, 0.03, 0.02, 0.9, 0.01], [0.01, 0.02, 0.01, 0.01, 0.8]]
S2 = [[0.9, 0.1, 0.05, 0.02, 0.01],[0.1, 0.8, 0.1, 0.02, 0.02], [0.05, 0.1, 0.7, 0.02, 0.01], [0.02, 0.02, 0.02, 0.6, 0.02], [0.01, 0.02, 0.01, 0.02, 0.7]]
```

```python
X1 = []
X2 = []
X1 = np.random.multivariate_normal(m1,S1,25).T
X2 = np.random.multivariate_normal(m2,S2,25).T
```

```python
mean1 = []
mean2 = []

for i in range(5):
    mean1.append(np.mean(X1[i]))

for i in range(5):
    mean2.append(np.mean(X2[i]))
```

```python

cov1 = []
cov2 = []

for i in range(5):
    cov1.append(np.cov(np.transpose(X1[i])))

for i in range(5):
    cov2.append(np.cov(np.transpose(X2[i])))
```

```python
x1 = []
x2 = []
x1 = np.random.multivariate_normal(mean1,S1,5000)
x2 = np.random.multivariate_normal(mean2,S2,5000)
```

```python
def nbc(x,cov,m):
    p = 1
    for j in range(5):
        p *= (1/(2*(math.pi)*(cov[j]))**(1/2))*(np.exp(-((x[j] - m[j])**2)/(2*cov[j])))
        return p 
```

```python
p1 = []
p2 = []

for i in range(5000):
    p1.append(nbc(x1[i],cov1,m1))
for i in range(5000):
    p1.append(nbc(x2[i],cov1,m1))   

for i in range(5000):
    p2.append(nbc(x1[i],cov2,m2))
for i in range(5000):
    p2.append(nbc(x2[i],cov2,m2))
```

```python
error1 = 0
error2 = 0
  

for i in range(10000):
    if(i<5000):
        if(p1[i]<p2[i]):
            error1+= 1
    else:
        if(p1[i]>p2[i]):
            error2+= 1
error_probability1= error1/5000 
error_probability2 =error2/5000
error_probability  = (error1+error2)/10000
print("Error probability for p1: ", error_probability1)
print("Error probability for p2:", error_probability2)
print("Error probability of the classifier: ", error_probability)

```

```python
fig = plt.figure(figsize = (10, 5))
plt.bar(["Error Probability in x1","Error Probability in x2","Total Error Probability"], [error_probability1, error_probability2, error_probability], width = 0.4)
plt.xlabel("Class")
plt.ylabel("Error Probability")
plt.show()
```
